```python
drop_column=['id','Unnamed: 32']
df.drop(drop_column,axis=1,inplace=True)
```

#### Encoding 

```python
df.diagnosis.unique()
```

```python
df['diagnosis']=df.diagnosis.replace({'M':1,'B':0})
```

```python
df.head(5)
```

```python
df['diagnosis']=df.diagnosis.astype("int")
```

### seperating the data into dependent and independent variable

```python
X=df.drop("diagnosis",axis=1)
y=df["diagnosis"]
```

```python
X
```

## normalizing/standardizing the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X
```

### spiliting the data into train and test and  training the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1234)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_test_predict=lr.predict(X_test)
```

```python
y_test_predict
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn import metrics as mt
```

```python
print(mt.confusion_matrix(y_test,y_test_predict))
```

```python
print(mt.classification_report(y_test,y_test_predict))
```

```python
grp=mt.confusion_matrix(y_test,y_test_predict)
```

```python
mt.ConfusionMatrixDisplay(grp).plot()
plt.title("Confusion_matrix")
```
